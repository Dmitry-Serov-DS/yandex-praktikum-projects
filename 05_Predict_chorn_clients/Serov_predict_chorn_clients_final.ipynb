## Постановка задачи

В моем распоряжении данные клиентов Банка «Бета-Банк», в которых указаны:  
- срок обслуживания,  
- страна проживания,  
- возраст,  
- кредитный рейтинг,  
- баланс на счете,  
- количество продуктов, которыми использует клиент,  
- наличие кредитной карты,  
- активность клиента,
- предполагаемая зарплата.

Стоит задача построить систему, способную проанализировать поведение клиентов и спрогнозировать уйдет ли клиент из Банка или нет.

## Исследование задачи

### Проверим дисбаланс классов

```python
target.value_counts(normalize=True).plot(kind='bar')
plt.xlabel('Ушедший клиент')
plt.xticks(rotation=0)
plt.ylabel('Доля клиентов')
plt.title('Распределение доли пользователей по признаку "ушедший"')
plt.show()
```

Сущесвует проблема дисбаланса

Создадим переменные, сохраним в них списки значений для дальшейшего подбора гиперпараметров модели через алгоритмы **RandomizedSearchCV** и **GridSearchCV**

```python
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 50, num = 5)]
max_depth = [int(x) for x in np.linspace(start = 1, stop = 15, num = 15)]
min_samples_split = [int(x) for x in np.linspace(start = 2, stop = 50, num = 10)]
min_samples_leaf = [int(x) for x in np.linspace(start = 2, stop = 50, num = 10)]
```

```python
#опредеим наилучшие параметры для модели
#создамим словарь из параметров, которые буду подбирать
parametrs = { 'n_estimators': n_estimators,
              'max_depth': max_depth,
              'max_features' : ['log2', 'sqrt'],
              'min_samples_leaf': min_samples_split,
              'min_samples_split': min_samples_leaf }
```

Создадим модель Случайного леса без учета баланса классов

```python
model = RandomForestClassifier(random_state=12345)
```

### RandomizedSearchCV

применю к этой модели алгоритм **RandomizedSearchCV** для оптимизации гиперпарамеров

```python
rs = RandomizedSearchCV(model, 
                        parametrs, 
                        n_iter = 100, 
                        cv = 3, 
                        verbose = 1, 
                        n_jobs=-1, 
                        random_state=12345)
```

```python
%%time
rs.fit(features_train, target_train);
```

Выведем получившиеся "лучшие" гиперпараметры в результате работы алгоритма **RandomizedSearchCV**

```python
rs.best_params_
```
